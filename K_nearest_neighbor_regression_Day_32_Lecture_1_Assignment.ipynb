In this assignment, we will learn about K nearest neighbor regression. We will use the absenteeism at work dataset loaded below and analyze the model generated for this dataset.

```python
%matplotlib inline

import math

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
absent = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/Absenteeism_at_work.csv', sep=';')
```

```python
absent.head()
```

Find which variables have the highest pairwise correlation and remove them from our dataset. Additionally, try to think of which variables are correlated by looking at the column names and remove those columns as well.

Note: When choosing between two categeoricla variables that are correlated, keep the one with fewer values.

```python
# answer below:
absent.corr().mean().sort_values(ascending=False) 
# sns.heatmap(absent.corr(), vmin=-1, vmax=1)

threshold = 0.75

for column in absent.columns:
  correlated = absent.corrwith(absent[column])
  correlated = correlated[abs(correlated) > threshold]
  if len(correlated) > 1:
    print("------------------")
    print("ColumnL {}".format(column))
    print(correlated)
```

```python
absent.Weight.shape #740
absent['Body mass index'].nunique() #17
absent['Weight'].nunique() #26

absent.drop(columns='Weight', inplace=True)
```

Find out which columns contain categorical variables and turn those into dummy variables.

```python
# answer below:
cat_cols = ['Reason for absence', 'Month of absence', 'Day of the week', 'Seasons', 'Education']
dummy_df = pd.get_dummies(absent, columns=cat_cols, drop_first=True, dtype=int)
dummy_df.head()
```

```python
Split the data into train and test with test containing 20% of the data.
```

```python
# answer below:
from sklearn.model_selection import train_test_split

X = dummy_df.drop(columns=['Absenteeism time in hours'])
y = dummy_df['Absenteeism time in hours']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() # only perform on training data
scaler.fit(X_train) # never use .fit() method on test
X_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

Train a KNN regression model using k=15 and compute the MSE for the test subsample.

```python
# answer below
from sklearn import neighbors

knn = neighbors.KNeighborsRegressor(n_neighbors=15)
knn.fit(X_scaled, y_train)
knn.score(X_test_scaled, y_test)

from sklearn.metrics import mean_squared_error 

y_pred = knn.predict(X_test_scaled)
mean_squared_error(y_pred, y_test) # MSE 114.139

# using scaled data, we produce a MSE of 89.39, which compared to the unscaled data's MSE of 114.139, the scaled data produces better results
```
